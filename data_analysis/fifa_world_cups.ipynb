# How did the group games end most often?

```python
cnts = wc14.loc[wc14.game_type == "group"]["score_home_vs_away"].value_counts()
freq = pd.concat([cnts, cnts/cnts.sum()], axis=1)
freq.columns = ["abs_freq", "rel_freq"]
freq
```

8 games of 48 games of the group stage, i.e. 16.6% ended on 2-1, while 5 games (10%) ended on 0-0.
